```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder         #from scikit-learn's preprocessing module.
                                                       #This is used to convert categorical text data into numerical labels,
                                                       # which is often required for machine learning algorithms.
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('/content/drive/MyDrive/CROP9.csv') # read the data set
data.head()
```

```python
input = data[['Season','N','P','K','Rainfall','Temperature','pH','Humidity','Soil_Type','Productivity','Crop']]
input.head() # display the input data
```

```python
LE = LabelEncoder() # calling labelencoder
```

```python
input['Season_n'] = LE.fit_transform(input['Season'])
input['Soil_n'] = LE.fit_transform(input['Soil_Type'])   #encoding all the categorical feature into numerical value
input['Region_n'] = LE.fit_transform(input['Soil_Type'])
input['Productivity_n']=LE.fit_transform(input['Productivity'])
input['Crop_n']=LE.fit_transform(input['Crop'])
```

```python
input
```

```python

 #.describe() method on the input DataFrame. This method generates descriptive statistics of the numerical columns in the DataFrame.
 input.describe()
```

```python
input.shape
```

```python
features = input[['K','pH','Humidity','Soil_n']] #input feature
features.shape
```

```python
target = input[['Productivity_n']] # output feature
target.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size= 0.30, random_state= 12)
X_test.shape   # split the dataset into trainig and testing part using train_test_split, where training size=70%,testing size=30%.
```

```python
#call decisiontree classifier also set the parameter
model = DecisionTreeClassifier(criterion='gini',splitter='best', max_depth=7,max_features='sqrt')
```

```python
model.fit(X_train, y_train) # fit the model
```

```python
# calculate the accuracy of the traing model
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
